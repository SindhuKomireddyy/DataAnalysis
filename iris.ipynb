# IRIS DATASETS

### 1. Import Libraries


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn theme
sns.set_theme(style="whitegrid")
```

### 2.Load Dataset

```python
# Load the Iris dataset from Seaborn
df = sns.load_dataset('iris')

# Display first few rows
df.head()
```

### 3.Explore Data

```python
print("Shape of dataset:", df.shape)
print("\nData Info:")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary Statistics:\n", df.describe())
```

### Insights:

### Dataset has 150 rows and 5 columns (4 numeric + 1 categorical).

### No missing values.

### Features are: sepal_length, sepal_width, petal_length, petal_width, and species.

### Each species has 50 samples.

### 4.Data Cleaning

```python
# Check for duplicates
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicates if any
df = df.drop_duplicates()
```

### Insights:

#### 1.No missing values or duplicates â€” dataset is clean and ready for analysis.

### 5.Analysis with NumPy & Pandas

```python
# Mean of each feature
mean_values = df.mean(numeric_only=True)
print("Average values of features:\n", mean_values)

# Average petal length by species
avg_petal_length = df.groupby('species')['petal_length'].mean()
print("\nAverage Petal Length by Species:\n", avg_petal_length)

# Average sepal width by species
avg_sepal_width = df.groupby('species')['sepal_width'].mean()
print("\nAverage Sepal Width by Species:\n", avg_sepal_width)
```

### Insights:

#### 1.Setosa has the smallest petals and widest sepals.

#### 2.Virginica has largest petals.

#### 3.Differences show how flower type can be identified from measurements.

### 6.Visualizations

1. Histogram of Petal Length
2. Boxplot of Sepal Width by Species
3. Scatter Plot: Petal Length vs Width

```python
sns.histplot(df['petal_length'], bins=15, kde=True, color='skyblue')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.show()
```

### Insight:

#### Petal lengths show 3 distinct clusters â€” one for each species.

```python
sns.boxplot(x='species', y='sepal_width', data=df, palette='Set2')
plt.title("Sepal Width by Species")
plt.show()
```

### Insight: 

#### Setosa generally has higher sepal width than others.

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.title("Petal Length vs Width by Species")
plt.show()
```

### Insight:

#### 1.Each species forms a distinct cluster.

#### 2.Perfect for classification problems.
