## Matplotlib-Python Visualization Library

### What is matplotlib?

matplotlib is a python library for creating static ,animated and interactive visualizations.

### Why we need it?

It helps to understand trends,patterns and outliers

### Where we use it?

1. DataAnalysis
2. Machine Learning(for plotting loss curves,linear regression etc)
3. Business purposes(analysing market trends, finance)
4. Scientific research

```python

 
 
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
 
plt.contour(X, Y, Z, cmap='plasma')
plt.title("Contour Plot Example")
plt.colorbar()
plt.show()
 
 
plt.contourf(X, Y, Z, cmap='viridis')
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
 
 
```

```python
from mpl_toolkits.mplot3d import Axes3D
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title("3D Surface Plot")
plt.show()
 
 
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
 
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')
fig.colorbar(surf)
ax.set_title("3D Surface with Colormap")
plt.show()
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue')
ax.set_title("3D Sphere Surface")
plt.show()
 
 
```

```python
# 1. Create a line plot of student marks across 5 exams
exams = [1, 2, 3, 4, 5]
marks = [78, 85, 92, 88, 76]

plt.plot(exams, marks, marker='o', color='blue', linestyle='-', linewidth=2)
plt.title("Student Marks Across 5 Exams")
plt.xlabel("Exam Number")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
# 2. Plot a bar chart of sales of 4 products
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [150, 200, 120, 180]

plt.bar(products, sales, color='green')
plt.title("Sales of Products")
plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.show()
```

```python
# 3. Make a scatter plot of Age vs Salary
age = [25, 30, 35, 40, 45, 50]
salary = [40000, 45000, 50000, 55000, 60000, 65000]

plt.scatter(age, salary, color='red', marker='x')
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

```

```python


# 4. Draw a histogram of marks scored by students in a class
marks_hist = [55, 65, 75, 85, 95, 70, 80, 90, 60, 75, 85]

plt.hist(marks_hist, bins=5, color='purple', edgecolor='black')
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()

```

```python
# 5. Customize a plot with different colors, markers, and line styles
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, color='orange', marker='s', linestyle='--', linewidth=2, markersize=8)
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 6. Create two subplots: Line plot and bar plot side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# Left subplot: Line plot
axes[0].plot(exams, marks, color='blue', marker='o', linestyle='-')
axes[0].set_title("Line Plot of Marks")
axes[0].set_xlabel("Exam")
axes[0].set_ylabel("Marks")

# Right subplot: Bar plot
axes[1].bar(products, sales, color='green')
axes[1].set_title("Sales Bar Chart")
axes[1].set_xlabel("Products")
axes[1].set_ylabel("Units Sold")

plt.tight_layout()
plt.show()
```
